#### Creating dictionary

```
dataF = pd.read_csv('/content/stacked-data-both-regions/data_stacked_T_ (1).csv') #  change path for false region if required
dataF = dataF[dataF['DTSM'].notna()]
dataF.shape 
```

```
well_logs  = pd.read_csv("/content/wellloginfo/WellLogsInfo.csv")
unit_list = []
l = list(dataF.columns)
unit = 'DEC'
for i in range(well_logs.shape[0]):
    k = well_logs.iloc[i,:]
    if (k[2] == unit and k[0] in l):
        unit_list.append(k[0])
        print("{} = {}".format(k[0],k[1]))
```

```
# Obsererving data distribution and info for selected values of a key
df = dataF[unit_list]
df = df.dropna(axis = 1, how = 'all')
df['DTSM'] = dataF.DTSM
df.info()
df.describe()
```

```
# Sorting Values in order of Correlation with DTSM
k = abs(df.corrwith(df['DTSM']))
k = k.sort_values(axis=0, ascending=False)
li_ = list(k.index)
li_.remove('DTSM')
print(li_)
```
