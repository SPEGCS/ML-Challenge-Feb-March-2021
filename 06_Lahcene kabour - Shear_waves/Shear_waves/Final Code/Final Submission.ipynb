```python

#### Read Files with 
file_location = "C:/Users/chebo/Desktop/Data Science/Gulf Section Datathon/Team Contributions/Test/Test CSV/"

List_of_Files = listdir(file_location)
Well_logs = []

#### Select CSV Files
for File in List_of_Files:
    if File[-3:]=='csv':
        Well_logs.append(File)

####
Logs=[]
for Well_log in Well_logs:
    Log = pd.read_csv(file_location + Well_log) 
    Logs.append(Log)


    
########        Extract Only One Gamma Ray column and save it into CSV
#
##
###
####
#####
######
def Save_Gamma(Logs,Well_logs):
    for Well_index,Log in enumerate(Logs):
                    
        Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
        one_log_columns=[]
        Column_index_list=[]
        for Column_number,column in enumerate(Log.columns):
      
            if 'GAMMA' in column:
                colu = column
                colu_index = Column_number
                Column_index_list.append(colu_index)
                one_log_columns.append(colu)
                
        if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:
            count_total_values_per_column=[]
            for i in Column_index_list:
                count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())
            
            max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
            max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
            del Column_index_list[max_values_gamma_index]
            
            Log.rename(columns={Log.columns[max_value_gamma_index_Final]:'GR'},inplace=True)
            Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)
           
        elif len(one_log_columns)  == 1 :
            Log.rename(columns={Log.columns[Column_index_list[0]]:'GR'},inplace=True)
            
       
        
Save_Gamma(Logs,Well_logs)      

########       
#
##
###
####
#####
###### Porosity
 
############## saving into two_columns from Gamma Columns Names that Has Only Two columns:
def Porosity_Data_Cleaning(Logs,T_D_N_S):
    
    
 
    if T_D_N_S == "THERMAL": ############## THERMAL
        SANDSTONE_LIMESTONE = ["'LIMESTONE' not in column and 'LS' not in column",
                               "'LIMESTONE' in column or 'LS' in column"]
        for formation_type in SANDSTONE_LIMESTONE:
            if "or" in formation_type:
                formation = "LIMESTONE"
            else:
                formation = "SANDSTONE"
            
            for Well_index,Log in enumerate(Logs):
                Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
        
                por_one_log_columns=[]
                por_Column_index_list=[]
                
                for Column_number,column in enumerate(Log.columns):
                  
                    if '.DEC' in column:
                       if 'THERMAL' in column: 
                           if "CROSSPLOT" not in column:
                               if formation_type:
                                       
                                    por_colu = column
                                    por_colu_index = Column_number
                                    por_Column_index_list.append(por_colu_index)
                                    
                                    por_one_log_columns.append(por_colu)
                if len(por_one_log_columns)  != 1 and (len(por_one_log_columns))!=0:
                    por_count_total_values_per_column=[]
                    for i in por_Column_index_list:
                        por_count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())
                    
                    por_max_values_gamma_index = por_count_total_values_per_column.index(max(por_count_total_values_per_column))
                    por_max_value_gamma_index_Final = por_Column_index_list[por_max_values_gamma_index]
                    del por_Column_index_list[por_max_values_gamma_index]
                    Log.rename(columns={Log.columns[por_max_value_gamma_index_Final]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)
                    Log.drop(Log.columns[por_Column_index_list],axis=1,inplace=True)
                elif len(por_one_log_columns)  == 1:
                    Log.rename(columns={Log.columns[por_colu_index]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)
                    
    
    ###### SONIC DENSITY NEUTRON
    
    
    else:
    
        
        SANDSTONE_LIMESTONE = ["'LIMESTONE' not in column and 'LS' not in column",
                           "'LIMESTONE' in column or 'LS' in column"]
        for formation_type in SANDSTONE_LIMESTONE:
            if "or" in formation_type:
                formation = "LIMESTONE"
            else:
                formation = "SANDSTONE"
        
        
    
        
    
    
    
            for Well_index,Log in enumerate(Logs):
                Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
        
                por_one_log_columns=[]
                por_Column_index_list=[]
                for Column_number,column in enumerate(Log.columns):
                  
                    if '.DEC' in column:
                       if 'THERMAL' not in column: 
                           if "CROSSPLOT" not in column:
                           
                               if T_D_N_S in column:
                               
                                                    
                                   if formation_type:
                                       
                                        por_colu = column
                                        por_colu_index = Column_number
                                        por_Column_index_list.append(por_colu_index)
                                        por_one_log_columns.append(por_colu)
                            
                if len(por_one_log_columns)  != 1 and (len(por_one_log_columns))!=0:
                    por_count_total_values_per_column=[]
                    for i in por_Column_index_list:
                        por_count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())
                    
                    por_max_values_gamma_index = por_count_total_values_per_column.index(max(por_count_total_values_per_column))
                    por_max_value_gamma_index_Final = por_Column_index_list[por_max_values_gamma_index]
                   
                    
                    del por_Column_index_list[por_max_values_gamma_index]
                    
                    Log.rename(columns={Log.columns[por_max_value_gamma_index_Final]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)
                    
                   
                    Log.drop(Log.columns[por_Column_index_list],axis=1,inplace=True)
                elif len(por_one_log_columns)  == 1:
                    Log.rename(columns={Log.columns[por_colu_index]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)
#####
#######
#########
###########       Run Poro Cleaning Function 
#########
#######
#####
                 
                 
                 
                 
Different_Por = ['NEUTRON','DENSITY','THERMAL','SONIC']
for Por_Type in Different_Por:
    
    Porosity_Data_Cleaning(Logs,Por_Type)                                
                                
#########
###########
############
#############
############## Density Data Cleaning
#############
############
###########
##########

def Save_Density(Logs,Well_logs,Density_Variables):
    
    ### Curve Name ###
    if any(case in "RHOB" for case in Density_Variables) == True:
        Curve_name = "Density.G/C3"
    elif any(case in "HDRA" for case in Density_Variables) == True:
        Curve_name = "Density.COR.G/C3"
    ###
        
    for Well_index,Log in enumerate(Logs):
                    
        Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
        one_log_columns=[]
        Column_index_list=[]
        for Column_number,column in enumerate(Log.columns):
      
            if '.G/C3' in column:
                if any(case in column for case in Density_Variables) == True:
                    colu = column
                    colu_index = Column_number
                    Column_index_list.append(colu_index)
                    one_log_columns.append(colu)
                
        if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:
            
            count_total_values_per_column=[]
            for i in Column_index_list:
                count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())
            
            max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
            max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
            del Column_index_list[max_values_gamma_index]
            
            Log.rename(columns={Log.columns[max_value_gamma_index_Final]:Curve_name},inplace=True)
            Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)
            
            
        elif len(one_log_columns)  == 1 :
            Log.rename(columns={Log.columns[Column_index_list[0]]:Curve_name},inplace=True)
            
            
Density_Variables = [['RHOB','RHOZ','ZDEN'],['HDRA','DRHO','RHOM','CORR.','DCOR.','DRH ','QRHO_','ZCOR']]    
for Variables in Density_Variables:
    
    Save_Density(Logs,Well_logs,Variables)     
    
####
#####
######
####### Cleaning Compressional Slowness
######
#####
####

def Save_Compressional_Slowness(Logs):
    for Well_index,Log in enumerate(Logs):
                    
        Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
        one_log_columns=[]
        Column_index_list=[]
        for Column_number,column in enumerate(Log.columns):
      
            
            if '.US/F' in column or '.MS/F' in column :
                if 'SHEAR' not in column and 'DTSM' not in column:
                    if 'DTCO.'  in column or 'DTCO ' in column:
                        colu = column
                        colu_index = Column_number
                        Column_index_list.append(colu_index)
                        one_log_columns.append(colu)
                        
                        
        
        if len(one_log_columns) > 0:  
            Log.rename(columns={Log.columns[Column_index_list[0]]:'CompSlowness.US/F'},inplace=True)
            
        
Save_Compressional_Slowness(Logs)     

###############
############
########
####
###### Resistivity OHMM Cleaning
####
########
##########
############

###
##
# Deep Resistivity
##
###


def Save_Resistivity_OHM_DEEP(Logs):
    for Well_index,Log in enumerate(Logs):
                    
        Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
        one_log_columns=[]
        Column_index_list=[]
        for Column_number,column in enumerate(Log.columns):
      
            if '.OHMM' in column:
                # if "RLA5" in column or "AT90" in column or "LLD" in column or "IDPH" in column:
                if any(case in column for case in ["RLA5","AT90","LLD","IDPH",'ILD']) == True:
                    
                    colu = column
                    colu_index = Column_number
                    Column_index_list.append(colu_index)
                    one_log_columns.append(colu)
                
        if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:
            count_total_values_per_column=[]
            for i in Column_index_list:
                count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())
            
            max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
            max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
            del Column_index_list[max_values_gamma_index]
            
            Log.rename(columns={Log.columns[max_value_gamma_index_Final]:'RESISTIVITY.DEEP.OHMM'},inplace=True)
            Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)
            
        elif len(one_log_columns)  == 1 :
            Log.rename(columns={Log.columns[Column_index_list[0]]:'RESISTIVITY.DEEP.OHMM'},inplace=True)
            
      

Save_Resistivity_OHM_DEEP(Logs)      
#
##
###
#### Shallow and Medium Resistivity 
###
##
#
Shallow_Resistivities = [["MSFL", "AT10","RLA1" ,"RXOZ"],
                         ["SFL","AT20","RLA2" ,"LLS"],'RESISTIVITY.SHALLOW.OHMM']
Medium_Resistivities =[["AT60" ,"RLA4","ILM","IMPH"],
                         ["RLA3","AT30"],'RESISTIVITY.MEDIUM.OHMM']


def Save_Resistivity_OHM_Shallow_Medium(Logs,Resistivities):
    for Well_index,Log in enumerate(Logs):
                    
        Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
        one_log_columns=[]
        second_log_columns=[]
        Column_index_list=[]
        second_Column_index_list=[]
        for Column_number,column in enumerate(Log.columns):
      
            if '.OHMM' in column:
                if any(case in column for case in Resistivities[0]) == True:
                    colu = column
                    colu_index = Column_number
                    Column_index_list.append(colu_index)
                    one_log_columns.append(colu)
                elif any(case in column for case in Resistivities[1]) == True: 
                    second_colu = column
                    second_colu_index = Column_number
                    second_Column_index_list.append(second_colu_index)
                    second_log_columns.append(second_colu)
                    
                
        if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:
            count_total_values_per_column=[]
            for i in Column_index_list:
                count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())
            
            max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
            max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
            del Column_index_list[max_values_gamma_index]
            
            Log.rename(columns={Log.columns[max_value_gamma_index_Final]:Resistivities[2]},inplace=True)
            Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)
            if len(second_Column_index_list)!=0:
                Log.drop(Log.columns[second_Column_index_list],axis=1,inplace=True)
            
            
        elif len(one_log_columns)  == 1 :
            Log.rename(columns={Log.columns[Column_index_list[0]]:Resistivities[2]},inplace=True)
            if len(second_Column_index_list)!=0:
                Log.drop(Log.columns[second_Column_index_list],axis=1,inplace=True)
            
        
        elif len(second_log_columns)  != 1 and (len(second_log_columns))!=0:
            count_total_values_per_column=[]
            for i in second_Column_index_list:
                count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())
            
            max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
            max_value_gamma_index_Final = second_Column_index_list[max_values_gamma_index]
            del second_Column_index_list[max_values_gamma_index]
            
            Log.rename(columns={Log.columns[max_value_gamma_index_Final]:Resistivities[2]},inplace=True)
            Log.drop(Log.columns[second_Column_index_list],axis=1,inplace=True)
            
        elif len(second_log_columns)  == 1 :
            Log.rename(columns={Log.columns[second_Column_index_list[0]]:Resistivities[2]},inplace=True)
            
            
      
Save_Resistivity_OHM_Shallow_Medium(Logs,Shallow_Resistivities)
Save_Resistivity_OHM_Shallow_Medium(Logs,Medium_Resistivities)
######
#######
########
#########
########## Keeping The columns in Question
#########
########
#######
######

def rename_DTSM(Logs):
    for Well_index,Log in enumerate(Logs):
               
        for Column_number,column in enumerate(Log.columns):
            if "DTSM" in column and "DTSM_" not in column :
                Log.rename(columns={Log.columns[Column_number]:"DTSM.US/F"},inplace=True)
            
rename_DTSM(Logs)        

######
#######
########
#########
########## Keeping The columns in Question
#########
########
#######
######
#    'THERMAL.SANDSTONE.DEC','SONIC.SANDSTONE.DEC','DENSITY.SANDSTONE.DEC','NEUTRON.SANDSTONE.DEC'
Final_Data_Columns =['RESISTIVITY.SHALLOW.OHMM',
                     'RESISTIVITY.MEDIUM.OHMM',
                     'RESISTIVITY.DEEP.OHMM',
                     'CompSlowness.US/F',
                     "Density.COR.G/C3",
                     "Density.G/C3",
                     'DENSITY.SANDSTONE.DEC',
                     'NEUTRON.SANDSTONE.DEC',
                     'NEUTRON.LIMESTONE.DEC',
                     'THERMAL.LIMESTONE.DEC',
                     'DENSITY.LIMESTONE.DEC',
                     'SONIC.LIMESTONE.DEC','GR',"DTSM.US/F"]
def Final_Cleaning(Logs):
    for Well_index,Log in enumerate(Logs):
        column_indices = []        
        for Column_number,column in enumerate(Log.columns):
          
            if all(case != column for case in Final_Data_Columns) ==True:
                column_indices.append(Column_number)
            
        if len(column_indices) !=0 :
            Log.drop(Log.columns[column_indices],axis=1,inplace=True)               

Final_Cleaning(Logs)
             
####              
######             
#########                
###########
##############
################ Adding All Data Into One File and removing NAN from DTSM 
##############
###########
########
######
####      


All_Data_One_File=pd.DataFrame(columns=Final_Data_Columns)

for Well_index,Log in enumerate(Logs):
    
    All_Data_One_File =All_Data_One_File.append(Log,ignore_index=True)


All_Data_One_File.dropna(subset=["DTSM.US/F"],inplace=True)    
All_Data_One_File.drop(columns = ["NEUTRON.SANDSTONE.DEC"],inplace=True)
All_Data_One_File.drop(columns = ['DENSITY.SANDSTONE.DEC'],inplace=True)


All_Data_One_File.to_csv(file_location + "FInal file.csv",index=False)

```

```python
##### Cleaning Submission Data
file_location = "C:/Users/chebo/Desktop/Data Science/Gulf Section Datathon/Submission/Final Submission Format/Team_Name/Wells/"   
Save_csv_file_location = "C:/Users/chebo/Desktop/Data Science/Gulf Section Datathon/Submission/Final Submission Format/Team_Name/Wells/CSV/"
Wells = listdir(file_location)

for count,Well in enumerate (Wells):
    count = count + 1 
    if Well[-3:] == "las":
        Log = pd.read_csv(file_location + Well,header=None, sep='\t',)
        for index,line in enumerate(Log[0]):
            if "Curve Information" in line:
                Upper = index+1
            elif   'Params' in line:
                Lower = index

        Columns = Log[Upper:Lower]
        Log_Data = Log[(Lower+3):]
        Log_Data = Log_Data[0].str.split(expand=True)

        Log_Data.columns = Columns
        
        Log_Data.to_csv(Save_csv_file_location+Well[:-4]+'.csv',index=False)

            
        
        ####
        
        Logs=[]
        Logs.append(Log_Data)
        
        ########### Cleaning Gamma Ray



        ########        Extract Only One Gamma Ray column and save it into CSV
        #
        ##
        ###
        ####
        #####
        ######
        def Save_Gamma(Logs):
            for Well_index,Log in enumerate(Logs):

                Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
                one_log_columns=[]
                Column_index_list=[]
                for Column_number,column in enumerate(Log.columns):
                    column = str(column)
                    if 'GAMMA' in str(column):
                        
                        
                        colu = column
                        colu_index = Column_number
                        Column_index_list.append(colu_index)
                        one_log_columns.append(colu)

                if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:
                    count_total_values_per_column=[]
                    for i in Column_index_list:
                        count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())

                    max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
                    max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
                    del Column_index_list[max_values_gamma_index]

                    Log.rename(columns={Log.columns[max_value_gamma_index_Final]:'GR'},inplace=True)
                    Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)

                elif len(one_log_columns)  == 1 :
                    Log.rename(columns={Log.columns[Column_index_list[0]]:'GR'},inplace=True)
            


        Save_Gamma(Logs)      
        
        ########       
        #
        ##
        ###
        ####
        #####
        ###### Porosity

        ############## saving into two_columns from Gamma Columns Names that Has Only Two columns:
        def Porosity_Data_Cleaning(Logs,T_D_N_S):



            if T_D_N_S == "THERMAL": ############## THERMAL
                SANDSTONE_LIMESTONE = ["'LIMESTONE' not in column and 'LS' not in column",
                                       "'LIMESTONE' in column or 'LS' in column"]
                for formation_type in SANDSTONE_LIMESTONE:
                    if "or" in formation_type:
                        formation = "LIMESTONE"
                    else:
                        formation = "SANDSTONE"

                    for Well_index,Log in enumerate(Logs):
                        Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                

                        por_one_log_columns=[]
                        por_Column_index_list=[]

                        for Column_number,column in enumerate(Log.columns):
                            column = str(column)
                            if '.DEC' in column:
                                if 'THERMAL' in column: 
                                    
                                    if "CROSSPLOT" not in column:
                                        if formation_type:
                                            
                                            por_colu = column
                                            por_colu_index = Column_number
                                            por_Column_index_list.append(por_colu_index)

                                            por_one_log_columns.append(por_colu)
                        if len(por_one_log_columns)  != 1 and (len(por_one_log_columns))!=0:
                            por_count_total_values_per_column=[]
                            for i in por_Column_index_list:
                                por_count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())

                            por_max_values_gamma_index = por_count_total_values_per_column.index(max(por_count_total_values_per_column))
                            por_max_value_gamma_index_Final = por_Column_index_list[por_max_values_gamma_index]
                            del por_Column_index_list[por_max_values_gamma_index]
                            Log.rename(columns={Log.columns[por_max_value_gamma_index_Final]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)
                            Log.drop(Log.columns[por_Column_index_list],axis=1,inplace=True)
                        elif len(por_one_log_columns)  == 1:
                            Log.rename(columns={Log.columns[por_colu_index]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)


            ###### SONIC DENSITY NEUTRON


            else:


                SANDSTONE_LIMESTONE = ["'LIMESTONE' not in column and 'LS' not in column",
                                   "'LIMESTONE' in column or 'LS' in column"]
                for formation_type in SANDSTONE_LIMESTONE:
                    if "or" in formation_type:
                        formation = "LIMESTONE"
                    else:
                        formation = "SANDSTONE"







                    for Well_index,Log in enumerate(Logs):
                        Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                

                        por_one_log_columns=[]
                        por_Column_index_list=[]
                        for Column_number,column in enumerate(Log.columns):
                            column = str(column)
                            if '.DEC' in column:
                                if 'THERMAL' not in column: 
                                    if "CROSSPLOT" not in column:

                                        if T_D_N_S in column:
                                            

                                            if formation_type:

                                                por_colu = column
                                                por_colu_index = Column_number
                                                por_Column_index_list.append(por_colu_index)
                                                por_one_log_columns.append(por_colu)

                        if len(por_one_log_columns)  != 1 and (len(por_one_log_columns))!=0:
                            por_count_total_values_per_column=[]
                            for i in por_Column_index_list:
                                por_count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())

                            por_max_values_gamma_index = por_count_total_values_per_column.index(max(por_count_total_values_per_column))
                            por_max_value_gamma_index_Final = por_Column_index_list[por_max_values_gamma_index]


                            del por_Column_index_list[por_max_values_gamma_index]

                            Log.rename(columns={Log.columns[por_max_value_gamma_index_Final]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)


                            Log.drop(Log.columns[por_Column_index_list],axis=1,inplace=True)
                        elif len(por_one_log_columns)  == 1:
                            Log.rename(columns={Log.columns[por_colu_index]:T_D_N_S +'.'+ formation+'.DEC'},inplace=True)
        #####
        #######
        #########
        ###########       Run Poro Cleaning Function 
        #########
        #######
        #####




        Different_Por = ['NEUTRON','DENSITY','THERMAL','SONIC']
        for Por_Type in Different_Por:

            Porosity_Data_Cleaning(Logs,Por_Type)                                

        #########
        ###########
        ############
        #############
        ############## Density Data Cleaning
        #############
        ############
        ###########
        ##########

        def Save_Density(Logs,Density_Variables):

            ### Curve Name ###
            if any(case in "RHOB" for case in Density_Variables) == True:
                Curve_name = "Density.G/C3"
                
            elif any(case in "HDRA" for case in Density_Variables) == True:
                Curve_name = "Density.COR.G/C3"
            ###

            for Well_index,Log in enumerate(Logs):

                Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
                one_log_columns=[]
                Column_index_list=[]
                for Column_number,column in enumerate(Log.columns):
                    column = str(column)
                    if '.G/C3' in column:
                        if any(case in column for case in Density_Variables) == True:
                            colu = column
                            colu_index = Column_number
                            Column_index_list.append(colu_index)
                            one_log_columns.append(colu)

                if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:

                    count_total_values_per_column=[]
                    for i in Column_index_list:
                        count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())

                    max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
                    max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
                    del Column_index_list[max_values_gamma_index]

                    Log.rename(columns={Log.columns[max_value_gamma_index_Final]:Curve_name},inplace=True)
                    Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)


                elif len(one_log_columns)  == 1 :
                    Log.rename(columns={Log.columns[Column_index_list[0]]:Curve_name},inplace=True)


        Density_Variables = [['RHOB','RHOZ','ZDEN'],['HDRA','DRHO','RHOM','CORR.','DCOR.','DRH ','QRHO_','ZCOR']]    
        for Variables in Density_Variables:

            Save_Density(Logs,Variables)     

        ####
        #####
        ######
        ####### Cleaning Compressional Slowness
        ######
        #####
        ####

        ########
        #########
        ##########
        ########### Save Compressionsal Slowness
        ##########
        #########
        ########

        def Save_Compressional_Slowness(Logs):
            for Well_index,Log in enumerate(Logs):

                Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
                one_log_columns=[]
                Column_index_list=[]
                for Column_number,column in enumerate(Log.columns):
                    column = str(column)

                    if '.US/F' in column or '.MS/F' in column :
                        if 'SHEAR' not in column and 'DTSM' not in column:
                            if 'DTCO.'  in column or 'DTCO ' in column:
                                colu = column
                                colu_index = Column_number
                                Column_index_list.append(colu_index)
                                one_log_columns.append(colu)



                if len(one_log_columns) > 0:  
                    Log.rename(columns={Log.columns[Column_index_list[0]]:'CompSlowness.US/F'},inplace=True)


        Save_Compressional_Slowness(Logs)     

        ###############
        ############
        ########
        ####
        ###### Resistivity OHMM Cleaning
        ####
        ########
        ##########
        ############

        ###
        ##
        # Deep Resistivity
        ##
        ###


        def Save_Resistivity_OHM_DEEP(Logs):
            for Well_index,Log in enumerate(Logs):

                Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
                one_log_columns=[]
                Column_index_list=[]
                for Column_number,column in enumerate(Log.columns):
                    column = str(column)
                    if '.OHMM' in column:
                        # if "RLA5" in column or "AT90" in column or "LLD" in column or "IDPH" in column:
                        if any(case in column for case in ["RLA5","AT90","LLD","IDPH",'ILD']) == True:

                            colu = column
                            colu_index = Column_number
                            Column_index_list.append(colu_index)
                            one_log_columns.append(colu)

                if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:
                    count_total_values_per_column=[]
                    for i in Column_index_list:
                        count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())

                    max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
                    max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
                    del Column_index_list[max_values_gamma_index]

                    Log.rename(columns={Log.columns[max_value_gamma_index_Final]:'RESISTIVITY.DEEP.OHMM'},inplace=True)
                    Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)

                elif len(one_log_columns)  == 1 :
                    Log.rename(columns={Log.columns[Column_index_list[0]]:'RESISTIVITY.DEEP.OHMM'},inplace=True)


        Save_Resistivity_OHM_DEEP(Logs)      
        #
        ##
        ###
        #### Shallow and Medium Resistivity 
        ###
        ##
        #
        Shallow_Resistivities = [["MSFL", "AT10","RLA1" ,"RXOZ"],
                                 ["SFL","AT20","RLA2" ,"LLS"],'RESISTIVITY.SHALLOW.OHMM']
        Medium_Resistivities =[["AT60" ,"RLA4","ILM","IMPH"],
                                 ["RLA3","AT30"],'RESISTIVITY.MEDIUM.OHMM']


        def Save_Resistivity_OHM_Shallow_Medium(Logs,Resistivities):
            for Well_index,Log in enumerate(Logs):

                Logs[Well_index].replace(-9999.25,np.nan,inplace=True)                
                one_log_columns=[]
                second_log_columns=[]
                Column_index_list=[]
                second_Column_index_list=[]
                for Column_number,column in enumerate(Log.columns):
                    column = str(column)
                    if '.OHMM' in column:
                        if any(case in column for case in Resistivities[0]) == True:
                            colu = column
                            colu_index = Column_number
                            Column_index_list.append(colu_index)
                            one_log_columns.append(colu)
                        elif any(case in column for case in Resistivities[1]) == True: 
                            second_colu = column
                            second_colu_index = Column_number
                            second_Column_index_list.append(second_colu_index)
                            second_log_columns.append(second_colu)


                if len(one_log_columns)  != 1 and (len(one_log_columns))!=0:
                    count_total_values_per_column=[]
                    for i in Column_index_list:
                        count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())

                    max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
                    max_value_gamma_index_Final = Column_index_list[max_values_gamma_index]
                    del Column_index_list[max_values_gamma_index]

                    Log.rename(columns={Log.columns[max_value_gamma_index_Final]:Resistivities[2]},inplace=True)
                    Log.drop(Log.columns[Column_index_list],axis=1,inplace=True)
                    if len(second_Column_index_list)!=0:
                        Log.drop(Log.columns[second_Column_index_list],axis=1,inplace=True)


                elif len(one_log_columns)  == 1 :
                    Log.rename(columns={Log.columns[Column_index_list[0]]:Resistivities[2]},inplace=True)
                    if len(second_Column_index_list)!=0:
                        Log.drop(Log.columns[second_Column_index_list],axis=1,inplace=True)


                elif len(second_log_columns)  != 1 and (len(second_log_columns))!=0:
                    count_total_values_per_column=[]
                    for i in second_Column_index_list:
                        count_total_values_per_column.append(Logs[Well_index].iloc[:,i].count())

                    max_values_gamma_index = count_total_values_per_column.index(max(count_total_values_per_column))
                    max_value_gamma_index_Final = second_Column_index_list[max_values_gamma_index]
                    del second_Column_index_list[max_values_gamma_index]

                    Log.rename(columns={Log.columns[max_value_gamma_index_Final]:Resistivities[2]},inplace=True)
                    Log.drop(Log.columns[second_Column_index_list],axis=1,inplace=True)

                elif len(second_log_columns)  == 1 :
                    Log.rename(columns={Log.columns[second_Column_index_list[0]]:Resistivities[2]},inplace=True)



        Save_Resistivity_OHM_Shallow_Medium(Logs,Shallow_Resistivities)
        Save_Resistivity_OHM_Shallow_Medium(Logs,Medium_Resistivities)
        ######
        #######
        ########
        #########
        ########## Keeping The columns in Question
        #########
        ########
        #######
        ######

        def rename_DTSM(Logs):
            for Well_index,Log in enumerate(Logs):

                for Column_number,column in enumerate(Log.columns):
                    column = str(column)
                    if "DTSM" in column and "DTSM_" not in column :
                        Log.rename(columns={Log.columns[Column_number]:"DTSM.US/F"},inplace=True)

        rename_DTSM(Logs)        

        ######
        #######
        ########
        #########
        ########## Keeping The columns in Question
        #########
        ########
        #######
        ######
        Final_Data_Columns =['RESISTIVITY.SHALLOW.OHMM',
                             'RESISTIVITY.MEDIUM.OHMM',
                             'RESISTIVITY.DEEP.OHMM',
                             'CompSlowness.US/F',
                             "Density.COR.G/C3",
                             "Density.G/C3",
                             'DENSITY.SANDSTONE.DEC',
                             'NEUTRON.SANDSTONE.DEC',
                             'NEUTRON.LIMESTONE.DEC',
                             'THERMAL.LIMESTONE.DEC',
                             'DENSITY.LIMESTONE.DEC',
                             'SONIC.LIMESTONE.DEC','GR',"DTSM.US/F"]
        
        def Final_Cleaning(Logs):
            for Well_index,Log in enumerate(Logs):
                column_indices = []       
                
                for Column_number,column in enumerate(Log.columns):
                    column = str(column)

                    if all(case != column for case in Final_Data_Columns) ==True:
                        column_indices.append(Column_number)

                if len(column_indices) !=0 :
                    Log.drop(Log.columns[column_indices],axis=1,inplace=True)               

        Final_Cleaning(Logs)

        ####              
        ######             
        #########                
        ###########
        ##############
        ################ Adding All Data Into One File and removing NAN from DTSM 
        ##############
        ###########
        ########
        ######
        ####      


        All_Data_One_File=pd.DataFrame(columns=Final_Data_Columns)

        for Well_index,Log in enumerate(Logs):
            
            All_Data_One_File =All_Data_One_File.append(Log,ignore_index=True)


        All_Data_One_File.drop(columns = ["NEUTRON.SANDSTONE.DEC"],inplace=True)
        All_Data_One_File.drop(columns = ['DENSITY.SANDSTONE.DEC'],inplace=True)
       
      
        All_Data_One_File.to_csv(Save_csv_file_location+Well[:-4]+'.csv',index=False)

    
```
